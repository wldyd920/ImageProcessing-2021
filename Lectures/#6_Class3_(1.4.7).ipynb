```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = cv2.imread('woman-hand-palm-25848097.jpg')
img2 = cv2.imread('Cartouche.jpg')
img3 = cv2.imread('Scarab-beetle.jpg')

img1 = img1.copy()
img2 = img2.copy()
img3 = img3.copy()
```

```python
imgray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,80,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
print(len(contours))
```

```python
img = cv2.drawContours(img1, contours, -1, (0, 255, 0), 3)
```

```python
plt.figure(figsize=(12,10))
plt.subplot(131),plt.imshow(imgray,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(thresh,cmap = 'gray')
plt.title('Threshold Image'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(img1,cmap = 'gray')
plt.title('Contour Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
