```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.png',0)

edges = cv2.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
edges1 = cv2.Canny(img,100,120) # 여기가 (,min,max) 파라미터
plt.subplot(121),plt.imshow(edges,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges1,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur = cv2.GaussianBlur(img,(3,3),0)
edges2 = cv2.Canny(blur,100,200)
plt.subplot(131),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(edges1,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(edges2,cmap = 'gray')
plt.title('Gaussian Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img2 = cv2.imread('fingerprint1.jpg')
blur2 = cv2.GaussianBlur(img2,(3,3),0)
edges3 = cv2.Canny(img2, 150, 300)
edges4 = cv2.Canny(blur2,150,300)
plt.figure(figsize=(12,10))
plt.subplot(131),plt.imshow(img2,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(edges3,cmap = 'gray')
plt.title('Gaussian Image'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(edges4,cmap = 'gray')
plt.title('Canny-only Image'), plt.xticks([]), plt.yticks([])
plt.show()
# morphological로 하는게 나음.
```
