```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('fingerprint2.jpg',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
kernel = np.ones((10,10),np.uint8)
erosion2 = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
```

```python
plt.figure(figsize=(8, 9))
plt.subplot(331), plt.imshow(img, 'gray'), plt.title('Input')
plt.subplot(332), plt.imshow(erosion, 'gray'), plt.title('erosion')
plt.subplot(333), plt.imshow(erosion2, 'gray'), plt.title('erosion2')
plt.subplot(334), plt.imshow(dilation, 'gray'), plt.title('dilation')
plt.subplot(335), plt.imshow(opening, 'gray'), plt.title('opening')
plt.subplot(336), plt.imshow(closing, 'gray'), plt.title('closing')
plt.subplot(337), plt.imshow(gradient, 'gray'), plt.title('gradient')
plt.subplot(338), plt.imshow(tophat, 'gray'), plt.title('tophat')
plt.subplot(339), plt.imshow(blackhat, ), plt.title('blackhat')
plt.show()

#  'gray'   .
```
