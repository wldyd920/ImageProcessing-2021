```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('letterR_opencv.png')

s = 3
kernel = np.ones((s,s),np.float32)/(s*s)
dst = cv2.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#  
blur = cv2.blur(img,(9,9))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#  
blur = cv2.GaussianBlur(img,(3,3),0) #   
plt.figure(figsize=(32,24))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# 
# sorting and choose middle if two, mean of them
s=5
median = cv2.medianBlur(img,s)
plt.figure(figsize=(32,24))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('sp2.png')
s = 5

blur = cv2.GaussianBlur(img,(s,s),0)
median = cv2.medianBlur(img,s)

plt.figure(figsize=(15,10))
plt.subplot(121),plt.imshow(blur),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()

# median is more powerful
```
