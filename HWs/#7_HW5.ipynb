```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = cv2.imread('sIMG_8253.jpg', cv2.IMREAD_COLOR)
img2 = cv2.imread('sIMG_8254.jpg', cv2.IMREAD_COLOR)
img3 = cv2.imread('sIMG_8255.jpg', cv2.IMREAD_COLOR)
img4 = cv2.imread('sIMG_8256.jpg', cv2.IMREAD_COLOR)
```

```python
img = img1
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
s = 15

blur = cv2.GaussianBlur(gray, (s,s), 0)
blur = cv2.medianBlur(blur,25)
blur = cv2.Canny(blur,30,135)
# th2 = cv2.adaptiveThreshold(blur, 200, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 5,11)
# kernel = np.ones((3, 3), np.uint8)
# closing = cv2.morphologyEx(th2, cv2.MORPH_CLOSE, kernel, iterations=4)
# cont_img = closing.copy()
# cont_img = th2.copy()
contours, hierarchy = cv2.findContours(blur, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

for i in range(len(contours)):
    cv2.drawContours(img, contours[i], -1, (0,255,0), 3)
    # add text
    x,y,w,h = cv2.boundingRect(contours[i])
    cv2.putText(img, 'C' + str(i+1),(x-10,y-10),cv2.FONT_HERSHEY_SIMPLEX,1.1,(255,0,0),2)

print("Number of Contours found = " + str(len(contours)))
```

```python
plt.figure(figsize=(24,16))
plt.subplot(221),plt.imshow(blur),plt.title('Canny')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(img),plt.title('Contour Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = img4
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
s = 31

blur = cv2.GaussianBlur(gray, (s,s), 0)
blur = cv2.medianBlur(blur, 21)
blur = cv2.Canny(blur,11,30)
th2 = cv2.adaptiveThreshold(blur, 200, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 5,11)
kernel = np.ones((3, 3), np.uint8)
closing = cv2.morphologyEx(th2, cv2.MORPH_CLOSE, kernel, iterations=4)
cont_img = closing.copy()
# cont_img = th2.copy()
contours, hierarchy = cv2.findContours(cont_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

for i in range(len(contours)):
    cv2.drawContours(img, contours[i], -1, (0,255,0), 3)
    # add text
    x,y,w,h = cv2.boundingRect(contours[i])
    cv2.putText(img, 'C' + str(i+1),(x-10,y-10),cv2.FONT_HERSHEY_SIMPLEX,1.1,(255,0,0),2)

print("Number of Contours found = " + str(len(contours)))

plt.figure(figsize=(24,16))
plt.subplot(221),plt.imshow(blur),plt.title('Canny')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(img),plt.title('Contour Image')
plt.xticks([]), plt.yticks([])
plt.show()
```
