```python
# Add Blur to original image
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = cv2.imread('woman-hand-palm-25848097.jpg')
img2 = cv2.imread('Cartouche.jpg')
img3 = cv2.imread('Scarab-beetle.jpg')
img4 = cv2.imread('1568374237-handmodel-3.jpg')
img5 = cv2.imread('Canopic-Jar.jpg')
img6 = cv2.imread('depositphotos_13473158-stock-photo-woman-hand.jpg')
img7 = cv2.imread('Djed.jpg')
img8 = cv2.imread('The-Ankh.jpg')

img1 = img1.copy()
img2 = img2.copy()
img3 = img3.copy()
img4 = img4.copy()
img5 = img5.copy()
img6 = img6.copy()
img7 = img7.copy()
img8 = img8.copy()
```

```python
img=img1
```

```python
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,80,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
blur = cv2.GaussianBlur(img,(3,3),0)
thresh = cv2.Canny(blur,100,200)
```

```python
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
print(len(contours))
```

```python
img = cv2.drawContours(img, contours, -1, (0, 255, 0), 3)
```

```python
plt.figure(figsize=(12,10))
plt.subplot(131),plt.imshow(imgray,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(thresh,cmap = 'gray')
plt.title('thresh Image'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(img,cmap = 'gray')
plt.title('Contour Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
