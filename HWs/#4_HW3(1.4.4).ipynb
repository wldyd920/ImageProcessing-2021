```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('lena.png')
```

```python
s = 15

blur = cv2.GaussianBlur(img,(s,s),0)
median = cv2.medianBlur(img,s)
kernel = np.ones((s,s),np.float32)/(s*s)
dst = cv2.filter2D(img,-1,kernel)

plt.figure(figsize=(15, 15))
plt.subplot(221),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(222),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))

plt.subplot(223),plt.imshow(blur),plt.title('Gaussian')
plt.xticks([]), plt.yticks([])
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))

plt.subplot(224),plt.imshow(median),plt.title('Median')
plt.xticks([]), plt.yticks([])
plt.imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))

plt.show()
```
