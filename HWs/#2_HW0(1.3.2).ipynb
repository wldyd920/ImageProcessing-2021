```python
# load image
img1 = cv2.imread('gray_image.jpg')
img2 = cv2.imread('image_2.jpg')

# check shape
print(img1.shape)
print(img2.shape)

#resize
img2.resize((194, 259, 3), refcheck=True)
print(img1.shape)

# add method 1
dst = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)

cv2.imshow('dst', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()

# add method 2
result = cv2.add(img1, img2)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()

# add method 3
result = img1 +img2
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
